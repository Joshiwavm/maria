```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nx = 1024
ny = 1024

x = np.arange(nx)
y = np.arange(ny)

kx = np.fft.fftfreq(nx)
ky = np.fft.fftfreq(ny)

KX, KY = np.meshgrid(kx, ky)

P = (1 + KX ** 2 + KY ** 2) ** (-11./12)
P[0, 0] = 0
```

```python
D = np.real(np.fft.ifft2(P * np.fft.fft2(np.random.standard_normal(size=(ny, nx)))))
```

```python
plt.imshow(D)
```
